```python
df['society']=df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★','', str(name)).strip()).str.lower()
```

```python
df['price'].dtypes
```

```python
df=df[df['price']!='Price on Request']
```

```python
def convert_value(x):
  if type(x)==float:
    return x
  else:
    if x[1]=='Lac':
       return round(float(x[0])*0.01,2)

    else:
      return round(float(x[0]),2)
```

```python
df['price']=df['price'].str.split(" ").apply(convert_value)
```

```python
df.isnull().sum()
```

```python
df['price_per_square_ft']=df['price_per_square_ft'].str.replace('₹','').str.replace(',','').str.replace('/sq.ft.','').str.strip().astype(float)
```

```python
df['bedRoom'].value_counts()
```

```python
df=df[~df['bedRoom'].isnull()]
```

```python
df['bedRoom']=df['bedRoom'].str.split(" ").str.get(0).str.strip().astype(int)
```

```python
df['bathroom'].value_counts()
```

```python
df[df['balcony'].isnull()]
```

```python
df['bathroom']=df['bathroom'].str.split(" ").str.get(0).str.strip().astype(int)
```

```python
df['balcony'].value_counts()
```

```python
df['balcony']=df['balcony'].str.split(" ").str.get(0).replace('No','0')
```

```python
df['additionalRoom'].value_counts()
```

```python
df['additionalRoom'].fillna('not available',inplace=True)
```

```python
df['additionalRoom']=df['additionalRoom'].str.lower()
```

```python
df['floorNum'].value_counts()
```

```python
type(df['floorNum'])
```

```python
def apply_regex(s):
  return re.sub(r'[a-z]','',str(s))
```

```python
df['floorNum']=df['floorNum'].str.split(" ").str.get(0).str.replace('Ground','0').str.replace('Basement','-1').str.replace('Lower','0').apply(apply_regex)
```

```python
df['facing']=df['facing'].fillna('NA')
```

```python
df['facing'].value_counts()
```

```python
df.insert(loc=4,column='Area',value=round((df['price']*10000000)/df['price_per_square_ft']))
```
