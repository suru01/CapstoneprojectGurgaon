```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
new_df=pd.read_csv('/content/houses (2).csv')
```

```python
new_df.head(10)
```

```python
new_df.info()
```

```python
new_df.insert(loc=3,column='Property_type',value='house')
```

```python
new_df.duplicated().sum()
```

```python
new_df=new_df.drop_duplicates()
```

```python
new_df.shape
```

```python
new_df.isnull().sum()
```

```python
new_df.drop(columns=['link','property_id'], inplace=True)
```

```python
new_df.head()
```

```python
new_df.rename(columns={'rate':'price_per_sqft'},inplace=True)
new_df.head()
```

```python
new_df['society'].value_counts()
```

```python
new_df['society'].value_counts().shape
```

```python
import re
new_df['society'] = new_df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
new_df['society'].value_counts().shape
```

```python
new_df['society'] = new_df['society'].str.replace('nan','independent')
```

```python
new_df.head()
```

```python
new_df['price'].value_counts()
```

```python
new_df = new_df[new_df['price'] != 'Price on Request']
```

```python
new_df.head()
```

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
new_df['price'] = new_df['price'].str.split(' ').apply(treat_price)
```

```python
new_df['price_per_sqft'].value_counts()
```

```python
new_df['price_per_sqft']=new_df['price_per_sqft'].str.split(" ").str.get(1).str.replace('/sq.ft.','').str.replace(',','').astype(float)
```

```python
new_df['bedRoom'].value_counts()
```

```python
new_df[new_df['bedRoom'].isnull()]
```

```python
new_df = new_df[~new_df['bedRoom'].isnull()]
```

```python
new_df.shape
```

```python
new_df['bedRoom'] = new_df['bedRoom'].str.split(' ').str.get(0).astype('int')
```

```python
new_df.head()
```

```python
new_df['bathroom'].value_counts()
```

```python
new_df['bathroom'].isnull().sum()
```

```python
new_df['bathroom'] = new_df['bathroom'].str.split(' ').str.get(0).astype('int')
```

```python
new_df.head()
```

```python
new_df['balcony'].value_counts()
```

```python
new_df['balcony'].isnull().sum()
```

```python
new_df['balcony'] = new_df['balcony'].str.split(' ').str.get(0).str.replace('No','0')
```

```python
new_df.head()
```

```python
new_df['additionalRoom'].value_counts()
```

```python
new_df['additionalRoom'].fillna('not available',inplace=True)
```

```python
new_df['additionalRoom'] = new_df['additionalRoom'].str.lower()
```

```python
new_df.head()
```

```python
new_df['noOfFloor'].value_counts()
```

```python
new_df['noOfFloor'].isnull().sum()
```

```python
new_df['noOfFloor'] = new_df['noOfFloor'].str.split(' ').str.get(0)
```

```python
new_df.head(20)
```

```python
new_df.rename(columns={'noOfFloor':'floorNum'},inplace=True)
```

```python
new_df['facing'].fillna('NA',inplace=True)
```

```python
new_df['area'] = round((new_df['price']*10000000)/new_df['price_per_sqft'])
```

```python
new_df.shape
```

```python
new_df.info()
```

```python
new_df.to_csv('houses_cleaned.csv',index=False)
```
